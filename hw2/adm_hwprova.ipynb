```python
import requests
from bs4 import BeautifulSoup
from string import ascii_lowercase
r={}
cnt = requests.get("http://www.bbc.co.uk/food/cuisines")
soup = BeautifulSoup(cnt.text , "lxml")
rl = {}
for link in soup.find_all("a"):
        if link.get("href").startswith('/food/cuisines') and not link.get('href')=='/food/cuisines':
            rl[link.get('href')] = ""
rectyp=[]
for key in rl:
        rectyp.append(key[15:])
print(rectyp)
for rt in rectyp:
        print("ricetta",rt)
        cnt = requests.get('http://www.bbc.co.uk/food/recipes/search?cuisines[]='+rt)
        soup = BeautifulSoup(cnt.text , "lxml")
        lst=[]
        for link in soup.find_all('a'):
            if link.get("href").startswith('/food/recipes/search?'):
                lst.append(link.contents[0])
        if lst:
            maxpage=int(lst[-2])
        else:
            maxpage=1
        i=0
        for page in range(1,maxpage+1):
            cnt = requests.get('http://www.bbc.co.uk/food/recipes/search?page='+str(page)+'&cuisines%5B0%5D='+rt+'&sortBy=lastModified')
            soup = BeautifulSoup(cnt.text , "lxml")
            for link in soup.find_all("a"):
                if(link.get("href").startswith('/food/recipes/')
                   and not link.get('href')=='/food/recipes/'
                   and "search" not in link.get('href')):
                    r[link.get('href')] = ""
                    i+=1
        print('fine',i)
```
